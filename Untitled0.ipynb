```python
sns.barplot(x="room_type", y=veri_df["room_type"].index, data=veri_df, color="blue")
plt.xticks(rotation=45)
plt.show()
#yazıların birbirine karısmaması için 45 derece döndürdüm
```

```python
sns.barplot(x="room_type", y="price", data=veri_df)
plt.title("Oda Türüne Göre Fiyat")
plt.xticks(rotation=45)
plt.show()
```

```python
# aylara ayırıp daha rahat bir çıktı almak istedim
veri_df["availability_month"] = pd.to_datetime(veri_df["availability_365"], origin='unix', unit='D').dt.month

sns.barplot(x="room_type", y="price", hue="availability_month", data=veri_df)
plt.title("Oda Türüne Göre Fiyat ve Aylara Göre Müsaitlik")
plt.xticks(rotation=45)
plt.show()
```

```python
# ortlama fiyatlarla inceleyelim
veri_df["availability_month"] = pd.to_datetime(veri_df["availability_365"], origin='unix', unit='D').dt.month
ortalama_fiyatlar = veri_df.groupby(['room_type', 'availability_month'])['price'].mean().reset_index()
sns.set_palette("pastel")
g = sns.catplot(x="availability_month", y="price", hue="room_type", data=ortalama_fiyatlar, kind="bar", ci=None, aspect=2.5)
g.set_axis_labels("Ay", "Ortalama Fiyat")
g.fig.suptitle("Oda Türüne ve Aya Göre Ortalama Fiyatlar", y=1.03)
plt.xticks(rotation=45)
plt.show()
```
